# Bay County Eviction Pressure with Disaster Segmentation

This notebook extends the early-warning framework by explicitly segmenting disaster, post-disaster, and recovery periods.


```python
import pandas as pd
import numpy as np

print('Bay County notebook loaded successfully')
```

```python
# Placeholder Bay County filings data (replace with Eviction Lab data)
df = pd.DataFrame({
    'year': list(range(2014, 2025)),
    'filings': [280, 290, 305, 320, 610, 590, 565, 545, 525, 515, 505]
})

# Disaster segmentation based on Hurricane Michael (2018)
df['period'] = np.where(df['year'] < 2018, 'pre_disaster',
                 np.where(df['year'] <= 2020, 'disaster_aftermath', 'recovery'))

df
```

```python
# Compare filings across periods
df.groupby('period')['filings'].agg(['count','mean','median','min','max']).reset_index()
```
